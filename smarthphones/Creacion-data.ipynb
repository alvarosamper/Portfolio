Creación de los directorios de las imágenes

```python
#Imports
import os
import random
```

```python
imagenes="/content/drive/MyDrive/movil-final/imagenes"
```

```python
def make_train_test(ntrain, dir_input):
  #DEFINIMOS DIRECTORIO DE ENTRADA
  dir_input= dir_input 
  input=[file for file in os.listdir(dir_input) if '.jpg' in file]

  #CONTAMOS LOS ARCHIVOS Y CREAMOS LA RELACION TRAIN-TEST
  ntest= 1-ntrain
  n=len(input)
  n_train= round(n*ntrain)
  n_test= round(n*ntest)
  #NOS ASEGURAMOS DE QUE NO SE PIERDA NINGUNA IMAGEN
  if n_train + n_test > n:
    n_test -=1
  elif n_train + n_test <n:
    n_train +=1

  #HACEMOS ALEATORIOS LOS DATOS Y LOS CONVERTIMOS EN BLANCO Y NEGRO
  random.shuffle(input)
  m=1
  print("Imágenes en train", n_train)
  print("Imágenes en test", n_test)

  #ESCRIBIMOS EL TRAIN
  with open('/content/drive/MyDrive/movil-final/imagenes/train.txt', 'w') as train_txt:
    # Going through all elements of the list
    for x in input[0:n_train]:
      train_txt.write(dir_input)
      train_txt.write("/")
      train_txt.write(x)
      train_txt.write("\n")

  with open("/content/drive/MyDrive/movil-final/imagenes/test.txt", "w") as test_txt:
    for x in input[n_train:]:
      test_txt.write(dir_input)
      test_txt.write("/")
      test_txt.write(x)
      test_txt.write("\n")
```

```python
make_train_test(0.85,"/content/drive/MyDrive/movil-final/imagenes")
```
