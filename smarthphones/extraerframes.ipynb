```python
#IMPORTS 
import cv2
import os
import datetime
```

```python
def extract_video(entrada, path, output):
  os.chdir(entrada)

  #RUTA DEL VÍDEO
  root= path

   #LEER CONTENIDO DEL VÍDEO
  cam= cv2.VideoCapture(root)
  print(cam)
  currentframe=0

  #SACAMOS INFORMACIÓN DEL VÍDEO
  frames= cam.get(cv2.CAP_PROP_FRAME_COUNT)
  fps= int(cam.get(cv2.CAP_PROP_FPS))
  seconds=int(frames/ fps)

  while True:
  #LEER FRAMES
    ret,frame=cam.read()
  
    if ret:
    #MIENTRAS HAYA FRAMES QEU EXTRAER
      nombre=path.replace(".mp4","")
      name= nombre +"frame"+str(currentframe) + ".jpg"

      if (currentframe % 3 ==0):

    #ESCRIBIR FRAME EXTRAIDO
        cv2.imwrite(output + name, frame)

      currentframe+=1
    else:
      
      return frames, fps, seconds
```

```python
entrada="/content/drive/MyDrive/movil-final/videos/entrenamiento"
output="/content/drive/MyDrive/movil-final/imagenes/sinetiquetar/"
```

```python
videos=[file for file in os.listdir(entrada) if '.mp4' in file]
```

```python
for x in videos:
  extract_video(entrada, x, output)
```
